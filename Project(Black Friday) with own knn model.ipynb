**LINEAR REGRESSION MODEL---**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X=df[['Product_Category_1','Product_Category_2','Product_Category_3']]
Y=df['Purchase']
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(X, Y, test_size = 0.3, random_state =42)
```

```python
lm=LinearRegression()
print(lm.fit(xTrain,yTrain))
print('Intercept=',lm.intercept_)
print('Coefficients=',lm.coef_)
```

**Here the right hand side column represents the change in the target variable when one unit of change is made to the respective dependent variable keeping all other factors constant..**

**For example if here we keep all other factors constant then an unit change in the product category 1 will change the purchase by 831 units....**

**KNN REGRESSION MODEL--**

```python
from sklearn.model_selection import train_test_split
x=df[["Product_Category_1",'Product_Category_2','Product_Category_3']]
y=df["Purchase"]
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.50,random_state = 42)
```

```python
from sklearn import neighbors
from sklearn.metrics import mean_squared_error 
from math import sqrt

rmse_val = [] 
rmse_val1=[]
for K in range(20):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)

    model.fit(x_train, y_train)  
    pred=model.predict(x_test) 
    error = sqrt(mean_squared_error(y_test,pred)) 
    error1=sqrt(mean_squared_error(y_train,pred))
    rmse_val.append(error)
    rmse_val1.append(error1)
    print('RMSE test value for k= ' , K , 'is:', error)
    print("RMSE train value for k=", K ,'is:', error1)
```

```python
print("The Minimum Test RMSE is=",min(rmse_val))
print("The Minimum Train RMSE is=",min(rmse_val1))
```

**TO KNOW THE K VALUE CORRESPONDING TO THE MINIMUM TEST AND TRAIN RMSE WE USE DICTIONARY IN PYTHON--- **

```python
d_test={}
d_train={}
for i in range(20):
    d_test.update({i+1:rmse_val[i]})
    d_train.update({i+1:rmse_val1[i]})
key_min=min(d_test.keys(),key=(lambda k:d_test[k]))    
key_min1=min(d_train.keys(),key=(lambda k:d_train[k]))
print("The minimum Test RMSE corresponding to the k value is=",key_min)   
print("The minimum Train RMSE corresponding to the k value is=",key_min1)    
    
```

```python
plt.plot(range(20),rmse_val,color='red')
plt.plot(range(20),rmse_val1,color='violet')
plt.xlabel("K")
plt.ylabel("RMSE")
plt.title("RMSE(TRAINING AND TESTING) VS K")
```

```python
f=[]
for h in range(1,21):
    p=1/h
    f.append(p)
    
plt.plot(f,rmse_val[0:20],color='blue')  
plt.plot(f,rmse_val1[0:20],color='green')
plt.xlabel("1/K")
plt.ylabel("RMSE")
plt.title("RMSE(TRAINING AND TESTING) VS 1/K")
```

**PREDICTING FOR THAT VALUE OF K WHICH HAVE THE LEAST RMSE-- **

```python
def predict(x_train,x_test,y_train,k):
    distances=[]
    targets=[]
    for i in range(len(x_train)):
        dist=np.sqrt(np.sum(np.square(x_train.values[i]-x_test)))
        distances.append((dist,y_train.values[i]))
        
    distances=sorted(distances)      
    p=0       
    
    for i in range(k):
        p=p+distances[i][1]
   
    return (p/k)
```
